```python
import pandas as pd
import numpy as np
from koeda import EDA
from koeda import AEDA
from tqdm import tqdm
import time
```

```python
data = pd.read_csv('/data/ephemeral/data/aug_data합친거_특문제거.csv')
```

```python
import re

def remove_unwanted_characters(text):
    # 한글, 영어, 숫자, 한자, 공백, 특정 특수문자 (. ·)만 남기고 제거
    return re.sub(r'[^가-힣a-zA-Z0-9\s\u4e00-\u9fff\.\·]', '', text)

# 데이터프레임의 text 열에 함수 적용
data['text'] = data['text'].apply(remove_unwanted_characters)
data.to_csv('dd.csv')
```

```python
data_bt = data.copy() 여기서 데이터 불러와서
```

```python
data_bt
```

```python
li=[]
for i, row in tqdm(data_bt.iterrows(), total=data_bt.shape[0], desc="Translating"):
    ID = row['ID']
    text = row['text']
    target = row['target']
    translated_text = GoogleTranslator(source='ko', target='en').translate(text)
    result = GoogleTranslator(source='en', target='ko').translate(translated_text)
    li.append(
        {
            'ID': ID,
            'text': result,
            'target': target
        }
    ) 
```

```python
li
```

```python
augmented_df = pd.DataFrame(li)
combined_data = pd.concat([data, augmented_df], ignore_index=True)
```

```python
combined_data.to_csv('BT리라벨링데이터증강.csv', index=False, encoding='utf-8-sig')
```

```python
full_data = pd.read_csv('/data/ephemeral/data/aug_data_특수문자제거.csv')
full_data = pd.concat([full_data, combined_data], ignore_index=True)
```
